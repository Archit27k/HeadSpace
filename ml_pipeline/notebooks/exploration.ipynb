# HeadSpace AI - Emotion Detection Data Exploration

This notebook is for exploring the emotion classification dataset, testing different text embeddings, and visualizing the data distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/emotions.csv')
df.head()
```

```python
# Plot label distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label', order=df['label'].value_counts().index)
plt.title('Distribution of Emotion Labels')
plt.show()
```
